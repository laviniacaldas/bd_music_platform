```sql
-- Consultando as 5 musicas mais ouvidas de um determinado ouvinte:

SELECT
    a.nome AS artista,
    COUNT(*) AS num_plays
FROM
    historico h
JOIN
    ouvintes o ON h.id_ouvinte = o.id_ouvinte
JOIN
    plays nt ON h.id_musica = nt.id_musica
JOIN
    artistas a ON nt.id_artista = a.id_artista
WHERE
    o.nome = 'João Silva' -- Escolhe o nome do ouvinte
GROUP BY
    a.nome
ORDER by -- ordena da mmais tocada pra menos tocada
    num_plays DESC
LIMIT 5; -- delimita as 5 primeiras musicas
```

```sql
-- Consultando as musicas mais tocadas de um determinado artista

SELECT
nome_musica AS nome,
COUNT(*) AS total_vezes_tocadas
FROM
    plays nt
JOIN
    artistas a ON nt.id_artista = a.id_artista
WHERE
    a.nome = 'Taylor Swift'
GROUP BY
    nome_musica
ORDER BY
    total_vezes_tocadas DESC
LIMIT 5;
```
